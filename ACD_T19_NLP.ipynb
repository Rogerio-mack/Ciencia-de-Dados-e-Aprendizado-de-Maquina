# **Tarefas** NLP
* `Tokenização` - divisão do texto em pedaços chamados tokens, que são significativos
unidades, como palavras e números.
* `Marcação de classes gramaticais (POS)` - identificando a classe gramatical de cada palavra, como
substantivo, verbo, adjetivo, etc.
* `Extração de substantivos` - localizando grupos de palavras que representam substantivos, como
“Fábrica de tijolos vermelhos”.
* `Análise de sentimento` - determinar se o texto tem positivo, neutro ou negativo
sentimento.
* `Tradução entre idiomas e detecção de idiomas` como a tecnologia do Google Translate.
* `Inflexão` - pluralização e singularização de palavras. 
* `Verificação ortográfica e correção ortográfica`.
* `Lematização` - como lematização, mas produz palavras reais com base no original
contexto das palavras. Por exemplo, a forma lematizada de "variedades" é "variedade".
* `Freqüências de palavras` - determinando com que freqüência cada palavra aparece em um corpus.
* `n-gramas` - produzindo conjuntos de palavras consecutivas em um corpus para uso na identificação de palavras que freqüentemente aparecem adjacentes umas às outras.

# A Biblioteca  [`spaCy`](https://spacy.io/usage/spacy-101) para tarefas um pouco mais sofisticadas
* `O que é a spaCy?`
spaCy é uma biblioteca de código aberto gratuita para Processamento de Linguagem Natural (PNL) avançado em Python. 

spaCy foi projetado especificamente para uso em produção e ajuda a criar aplicativos que processam e “entendem” grandes volumes de texto. Ele pode ser usado para construir sistemas de extração de informações ou de compreensão de linguagem natural, ou para pré-processar texto para aprendizado profundo.

* `O que spaCy não é?`
spaCy não é uma plataforma ou “uma API”. Ao contrário de uma plataforma, spaCy não fornece um software como serviço ou aplicativo da web. É uma biblioteca de código aberto projetada para ajudá-lo a construir aplicativos de PNL, não um serviço consumível.


spaCy não é um mecanismo de bot de bate-papo pronto para usar. Embora spaCy possa ser usado para alimentar aplicativos de conversação, não foi projetado especificamente para bots de bate-papo e fornece apenas os recursos de processamento de texto subjacentes.




## Tarefas com a spaCy

### spaCy’s Statistical Models

Esses modelos são os motores da spaCy. Esses modelos permitem que a spaCy execute várias tarefas relacionadas ao PNL, como TAGGING, POS, NER etc.

Os diferentes modelos estatísticos em spaCy, juntamente com suas especificações são:

    en_core_web_sm: English multi-task CNN trained on OntoNotes. Size – 11 MB
    en_core_web_md: English multi-task CNN trained on OntoNotes, with GloVe vectors trained on Common Crawl. Size – 91 MB
    en_core_web_lg: English multi-task CNN trained on OntoNotes, with GloVe vectors trained on Common Crawl. Size – 789 MB

Para a maior parte das tarefas que veremos aqui podemos usar o _sm, ao menos que indicado o contrário. Isso nos ajuda a economizar espaço e tempo de processamento sacrificando um pouco da precisão. A seguior veremos como realizar algumas das tarefas. 


## Tokenização
Durante o processamento, spaCy primeiro transforma o texto em tokens, ou seja, o segmenta em palavras, pontuação e assim por diante. Isso é feito aplicando regras específicas para cada idioma. Cada documento consiste em tokens individuais e podemos iterar sobre eles como no exemplo abaixo:

```python
import spacy
nlp = spacy.load("en_core_web_sm") #observe aqui o modelo sm
doc = nlp("This is a sentence.") 
print(doc, type(doc))

#iterando cada elemento da frase em inglês acima.
for token in doc:
  print(token)
```

### O que aconteceu acima?

* Depois de importar o módulo spacy na célula acima, carregamos um modelo e o chamamos de nlp.
* Em seguida, criamos um objeto Doc aplicando o modelo ao nosso texto e o nomeamos doc.
* O objeto Doc que contém o texto processado.

```python
# Da mesma forma, podemos tokenizar por sentenças também.
nlp = spacy.load("en_core_web_sm") 
doc = nlp("This is a sentence.This is another sentence")
doc

#iterando cada elemento da frase em inglês acima.
for i in doc.sents:
  print(i)
```

```python
#tokenização por chunks (noun chunks)
#doc2 = nlp("tesla is a automobile based endorsed with high tech work for implimenting the electric cars")
for chunks in doc.noun_chunks:
  print(chunks)
```

##Lematização
Um lema é a forma básica de um token. Você pode pensar nisso como a forma em que o token apareceria se estivesse listado em um dicionário. Por exemplo, o lema para o token "voando" é "voar". Lematização é o processo de reduzir as formas das palavras ao seu lema. O script a seguir fornece um exemplo simples de como fazer lematização com spaCy:

```python
#lematizando cada elemento da frase em inglês acima.
doc = nlp("This is a sentence.")
doc

for token in doc:
  print(token.text,token.lemma_)
```

Observe na expressão acima que is se tornou be, que é a forma original do verbo conjugado. 

No geral estamos interessados nos elementos do texto e na função destes elementos em uma frase, e não apenas na lista de elementos. 

## Part-of-Speech Tagging (PoS)
Uma tag de classe gramatical informa a classe gramatical (substantivo, verbo e assim por diante) de uma determinada palavra em uma determinada frase.

```python
doc = nlp("Apple is looking at buying U.K. startup for $1 billion")

for token in doc:
    print(token.text, token.lemma_,token.tag_)
```

A lista completa de tags pode ser encontrada [aqui](https://https://web.archive.org/web/20190206204307/https://www.clips.uantwerpen.be/pages/mbsp-tags). Seguem alguns exemplos abaixo:

## Caracterizações importantes do texto

    Text: o texto da palavra original.
    Lemma: a forma básica da palavra.
    POS: A tag de classe gramatical UPOS simples.
    Tag: a tag detalhada da classe gramatical.
    Dep: dependência sintática, ou seja, a relação entre tokens.
    Shape: a forma da palavra - letras maiúsculas, pontuação, dígitos.
    is alpha: o token é um caractere alfa?
    is stop: o token faz parte de uma lista de stop words, ou seja, as palavras mais comuns do idioma.

```python
doc = nlp("Apple is looking at buying U.K. startup for $1 billion")

for token in doc:
    print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
            token.shape_, token.is_alpha, token.is_stop)
```

# Vetorizando Texto (Word Embeddings)

Vetores de palavras - também chamados de *embeddings de palavras* - são descrições matemáticas de palavras individuais, de forma que palavras que aparecem frequentemente juntas no idioma tenham valores semelhantes. Desta forma, podemos derivar matematicamente o *contexto*.

```python
nlp(u'cat').vector
```

Para entender o que está acontecendo abaixo vamos utilizar o exemplo dado na seção 6.3 de https://web.stanford.edu/~jurafsky/slp3/6.pdf . O exemplo pode ser melhor entendido a partir das seguintes figuras abaixo:

Note que no modelo acima um documento é representado a partir de um vetor de contagem de palavras. 

Os vetores das comédias As You
Like It [1.114,36,20] e Twelfth Night [0,80,58,15] se parecem muito mais
(mais fool e wit do que batles) do que se parecem com Júlio César [7,62,1,2] ou
Henry V [13,89,4,3]. Isso fica claro com os números brutos; na primeira dimensão
(batalha) as comédias têm números baixos e as outras têm números altos.

Uma matriz de documento de termo real, é claro, não teria apenas 4 linhas e colunas,
quanto mais 2. De maneira mais geral, a matriz termo-documento tem | V | linhas (uma para cada
tipo de palavra no vocabulário) e D colunas (uma para cada documento na coleção
ção). Os tamanhos do vocabulário são geralmente na casa das dezenas de milhares, e o
o número de documentos pode ser enorme (pense em todas as páginas da web por exemplo).

Para documentos, vimos que documentos semelhantes tinham vetores semelhantes, porque documentos semelhantes tendem a ter palavras semelhantes. Este mesmo princípio se aplica às palavras:
palavras semelhantes têm vetores semelhantes porque tendem a ocorrer em documentos semelhantes.
A matriz termo-documento, portanto, nos permite representar o significado de uma palavra pelo documento. 

Uma alternativa ao uso da matriz termo-documento para representar palavras como vetores de
contagem de documentos, é usar a matriz termo-termo, também chamada de matriz palavra-palavra ou a matriz de contexto-termo, na qual as colunas são rotuladas por palavras, em vez de documentos.

O contexto pode ser o documento, em
caso em que a célula representa o número de vezes que as duas palavras aparecem na mesma
documento. É mais comum, no entanto, usar contextos menores, geralmente uma solução
percorrer a palavra, por exemplo, 4 palavras à esquerda e 4 palavras à direita,
nesse caso, a célula representa o número de vezes (em algum corpus de treinamento) o
a palavra da coluna ocorre em uma janela de ± 4 palavras ao redor da palavra da linha (n-grams).

# O Algoritmo TF-IDF

Pense sobre o seguinte argumento:

**Palavras que ocorrem mais frequentemente são mais importantes do que palavras que ocorrem uma ou duas vezes. Por outro lado, palavras que são muito frequentes são pouco importantes.**

Uma solução para a assertiva acima é encontrar um ponto de equilíbrio. É exatamente isso que o algoritmo tf-idf faz.

$$tf_{t,d}=log(count(t,d)+1)$$
$$idf_{t}=log\left(\frac{N}{df_{t}}\right)$$
$$w_{t,d}=tf_{t,d} \times idf_{t}$$


Aqui estão alguns valores idf para algumas palavras no corpus de Shakespeare, variando de
palavras extremamente informativas que ocorrem em apenas uma peça como `Romeu`, ou aquelas que
ocorrem em algumas como `salad` ou `Falstaff`, para aqueles que são muito comuns como `wit` ou ainda as tão comuns a serem completamente inúteis para discriminar, uma vez que ocorrem em todas as 37 peças, como `good` ou `sweet`.
