# O que  uma srie temporal?

> **A *Srie Temporal*  uma sequncia de observaes registradas em intervalos de tempo regulares.**

Dependendo da frequncia das observaes, uma srie temporal pode ser horria, diria, semanal, mensal, trimestral e anual, mas tambm pode haver sries temporais de segundos e minutos, como nmero de cliques de visitas em um site; Com segundos ou milisegundos, como para algumas sries encontradas em fsica; Ou mesmo dcadas e perodos mais longos como encontramos em astronomia e geologia. So muitos os exemplos de dados que temos interesse em observar no tempo: 

* O PIB de um pas ao longo dos anos
* Produo de safras anuais agrcolas
* Preos dirios de aes e commodities 
* Evoluo das temperatura da terra e das emisses de CO2
* Vendas e demandas de produtos de uma empresa ao longo do ano 
* Aumento do nmero de funcionrios ao longo do ano
* Reduo do nmero de fumantes
* Aumento de casos de uma doena ou evoluo do tempo de reestabelecimento dos pacientes
* etc.

e, em geral, estamos interessados em **analisar esses dados para predies e tomada de deciso.**

## Modelo Ingnuo de Mdias

Talvez, sem conhecer nada de uma srie temporal, uma primeira estimativa poderia ser o seu valor mdio ao longo do tempo. Esse valor  til, pois um modelo *ingnuo* pode servir de base para modelos mais elaborados. 

Na mesma linha podemos ainda refinar essa estimativa avaliando os valores mdios para diferentes perodos da srie e, se voc pensar que em muitos casos valores mais recentes podem ter uma influncia maior na previso de novos valores, faz sentido observarmos a mdia de valores mais recentes. 

```python
def plot_means(df,ax,t):
  ix = np.linspace(0,len(df),5).astype(int)
  ax.bar(range(4), np.array([df.reset_index().iloc[ ix[0]:ix[1] ].value.mean(),
                              df.reset_index().iloc[ ix[1]:ix[2] ].value.mean(),
                              df.reset_index().iloc[ ix[2]:ix[3] ].value.mean(),
                              df.reset_index().iloc[ ix[3]:ix[4] ].value.mean()]))
  ax.hlines(df.value.mean(),0,5,color='r',linestyle='--')
  ax.text(0, df.value.mean(), 'Overall mean')
  ax.set_title('Quarter Means ')

  return

fig, ax = plt.subplots(1,3, figsize=(20,4))

plot_means(only_trend, ax=ax[0],t='Trend Only')
plot_means(only_season, ax=ax[1],t='Season Only')
plot_means(trend_and_season, ax=ax[2],t='Trend and Season')

plt.tight_layout()
plt.show()
```

```python
def plot_means2(df,ax,t):
  ix = np.linspace(0,len(df),5).astype(int)
  df.plot(title=t, legend=False, ax=ax)
  ax.hlines(df.reset_index().iloc[ ix[3]:ix[4] ].value.mean(), df.index[0], df.index[-1],color='r',linestyle='--')
  ax.text(df.index[0], df.reset_index().iloc[ ix[3]:ix[4] ].value.mean(), 'Recent mean')

  ax.hlines(df.reset_index().value.mean(), df.index[0], df.index[-1],color='b',linestyle='--')
  ax.text(df.index[0], df.reset_index().value.mean(), 'Overall mean')

  return

fig, ax = plt.subplots(1,3, figsize=(20,4))

plot_means2(only_trend, ax=ax[0],t='Trend Only')
plot_means2(only_season, ax=ax[1],t='Season Only')
plot_means2(trend_and_season, ax=ax[2],t='Trend and Season')

plt.tight_layout()
plt.show()
```

## Modelo ingnuo de Regresso

Alguns tipos de sries, como sries onde predomina a tendncia, um modelo de regresso pode ser til. Mas note que o modelo de regresso no  um modelo temporal.

> **O modelo de regresso no  exatamente um modelo temporal! Note que voc no pode escrever um modelo $f(t) = a_0 + a_1 t$ onde $t$  o tempo da forma que empregamos nas sries anteriores YYYY-MM-DD ou algo assim.**

Podemos entretanto criar criar uma coluna de *time-steps*, ou perodos, que podemos ento empregar diretamente. 

$$ \hat y = a_0 + a_1 \text{time_steps} $$

# Caso: **Estimando a emisso de CO2 no Mundo** 

Vamos empregar modelos de regresso simples para estimar as emisses de CO2 a partir da srie histrica de emisses de gases para os prximos 36 meses.

Dados: http://meusite.mackenzie.br/rogerio/data_load/co2.csv

```python
df = pd.read_csv('http://meusite.mackenzie.br/rogerio/data_load/co2.csv')

df.index = pd.to_datetime(df['Date'], format='%Y-%m-%d')
df.drop(columns='Date',inplace=True)
df.head()
```

```python
plt.figure(figsize=(12,3))
plt.plot(df.CO2)
plt.title('Emisses de CO2')
plt.show()
```

Por simplicidade vamos chamar nossa coluna de time. 

```python
df['time'] = np.arange(len(df))
df.head()
```

```python
plt.figure(figsize=(12,3))
sns.regplot(x='time', y='CO2', data=df, scatter_kws=dict(color='0.6'), line_kws=dict(color='red'))

plt.show()
```

### Modelo de Regresso

```python
model = sm.ols(formula='CO2 ~ time', data=df)
result = model.fit()
print(result.summary())
```

### Predict 36 meses a frente... 

```python
new_times = pd.DataFrame()
new_times['time'] = np.arange(len(df), len(df) + 36)

new_times['prediction'] = result.predict(new_times)
new_times.head()
```

```python
plt.figure(figsize=(15,4))
sns.regplot(x='time', y='CO2', data=df, scatter_kws=dict(color='0.6'), line_kws=dict(color='blue'))

sns.lineplot(x='time', y='prediction', data=new_times, color='red')

plt.xticks(ticks=[ df['time'][0], df['time'][len(df)-1] ],labels=[ df.index[0], df.index[-1] ])
plt.show()
```

E esse modelo, certamente  uma estimativa bastante razovel do aumento de emisses para os prximos 3 anos.

# Decomposio de Sries Temporais

```python
#@markdown plot_series
def plot_series(values, decomp_type='additive', max_lags=30):
  #
  # values = a pandas series with datetime index
  #
  import warnings
  warnings.filterwarnings('ignore')

  import numpy as np
  import pandas as pd
  import matplotlib.pyplot as plt
  %matplotlib inline 
  import seaborn as sns
  import statsmodels.formula.api as sm

  from statsmodels.graphics.tsaplots import plot_acf
  from statsmodels.graphics.tsaplots import plot_pacf
  from statsmodels.tsa.stattools import adfuller
  from statsmodels.tsa.seasonal import seasonal_decompose
  from statsmodels.tsa.stattools import periodogram

  from pylab import rcParams
  rcParams['figure.figsize'] = 17,3

  plt.plot(values.index, values, c='r')
  plt.title('Original | Observed Series')
  plt.show()

  rcParams['figure.figsize'] = 14,3
  fig, ax = plt.subplots(nrows=1, ncols=3, sharey=True)

  decomposition = seasonal_decompose(values, model=decomp_type)

  ax[0].plot(decomposition.trend)
  ax[0].set_title('Trend')
  ax[1].plot(decomposition.seasonal)
  ax[1].set_title('Seasonal')
  ax[2].plot(decomposition.resid)
  ax[2].set_title('Residuals')

  plt.tight_layout()
  plt.show()

  rcParams['figure.figsize'] = 14,3
  fig, ax = plt.subplots(nrows=1, ncols=3)

  plot_acf(values, ax=ax[0], lags = max_lags)
  plot_pacf(values, ax=ax[1], lags = max_lags)

  periods = periodogram(values)
  # print(periods)
  ax[2].bar(x=range(len(periods)), height=periods)
  plt.title('Periodogram')

  plt.tight_layout()
  plt.show()

  warnings.filterwarnings('once')

  return 


```

Uma **tendncia**  observada quando h uma inclinao crescente ou decrescente observada na srie temporal. J a **sazonalidade**  observada quando h um padro repetido distinto observado entre intervalos regulares devido a fatores sazonais. Pode ser por causa do ms do ano, do dia do ms, dos dias da semana ou mesmo da hora do dia. Mas, como vimos antes, no  obrigatrio que todas as sries temporais tenham tendncia e/ou sazonalidade. Uma srie temporal pode no ter uma tendncia distinta, mas ter uma sazonalidade e o contrrio tambm pode acontecer. H ainda uma componente de **erro**. 

## Sries temporais aditivas e multiplicativas

Dependendo da natureza da tendncia e da sazonalidade, uma srie temporal pode ser modelada como aditiva ou multiplicativa, em que cada observao da srie pode ser expressa como uma soma ou um produto dos componentes:

* Srie temporal aditiva:

$$Valor = Tendncia +  Sazonalidade + Erro$$

* Sries Temporais Multiplicativas:

$$ Valor = Tendncia \times Sazonalidade \times Erro$$

# Sries no Estacionrias

Uma srie no estacionria  uma srie que tem suas estatsticas alteradas ao longo do tempo.

> **Uma srie no estacionria  uma srie que tem suas estatsticas alteradas ao longo do tempo. Por exemplo, a mdia e a varincia, e assim tambm o desvio padro, mudam significativamente ao longo do tempo.**

Isso  um grande um problema para as predies e uma forma de contornarmos isso  fazendo transformaes das sries temporais tornando-as estacionrias. So comuns transformaes do tipo $log(x_t)$ ou de potncias $x_t^2$, mas antes disso vamos ver um teste para verificar se a uma srie , de fato, signficativamente estacionria ou no.

# ADF Augmented Dickey-Fuller 

O teste **ADF Augmented Dickey-Fuller**  um teste para verificar se uma srie temporal  ou no estacionria de modo significativo.  um dos testes mais simples e voc pode buscar outros testes mais elaborados se precisar aplic-los de fato. Na prtica, todos esses testes so limitados a poucas caractersticas de no estacionariadade, e pode ser necessrio aplicar mais de um teste e, mesmo, uma inspeo visual pode ser s vezes mais importante e reveladora.

Aqui a hiptese nula do teste  que a srie temporal pode ser representada por uma raiz unitria, que no  estacionria (tem alguma estrutura dependente do tempo). A hiptese alternativa (rejeitando a hiptese nula)  que a srie temporal  estacionria.

Assim, de modo prtico,

$p-valor \gt 0,05$ : Falha ao rejeitar a hiptese nula ($H0$), **a srie no  estacionria**.

$p-value \le 0.05$ : Rejeitar a hiptese nula ($H0$), **a srie  estacionria**.

```python
import ipywidgets as widgets

def f(P=20,noise=0.5,trend=4):
  index_dates = pd.date_range(start='1/1/2000', end='1/1/2022', freq='M')
  values = np.sin( np.arange(0,len(index_dates),1) * np.pi/P )  +  np.cos( np.arange(0,len(index_dates),1) * np.pi/(2*P) ) + 2
  values = values * noise*np.random.sample(len(index_dates))
  values = values * trend*np.linspace(0,1,len(index_dates)) + 1
  df = pd.DataFrame()
  df.index = index_dates 
  df['values'] = values

  plot_series(df['values'])  

  return df

w = widgets.interactive(f, P=(1, 100, 1), noise=(0, 1, 0.01), trend=(0,20,0.1))
display(w)

w.result
```

```python
from statsmodels.tsa.stattools import adfuller

df = w.result

X = df['values']
result = adfuller(X)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
	print('\t%s: %.3f' % (key, value))
```

# Tornando uma Srie Estacionria 

Vrias transformaes podem ser aplicadas para tornar uma srie estacionria. A ideia  aplicar a transformao, empregar a srie estacionria para a criar um modelo de predio e, feita a predio, aplicar a inversa da transformao para obter o resultado da srie original.

Vrios tipos de transformao podem ser aplicadas o que vai depender da srie de dados e no uma regra. Mas so comuns a aplicao de $log(x_t)$, funes de potncia como $x_t^n$ ou ainda funes de diferena, como abaixo, que podem remover tendncias e sazonalidades dos dados.





## Transformao de diferena
A diferenciao  um mtodo de transformar um conjunto de dados de srie temporal que pode tanto remover tendncias e sazonalidades dos dados.

A ideia transformarmos a srie de dados em uma srie de diferenas em um dado intervalo para obtermos a srie estacionria $y_e$.

$$ y_e(t) = y(t) - y(t- \text{intervalo}) $$

Vamos criar uma nova srie $y_e$ baseada na srie original $y$ considerando o intervalo $1$ do seguinte modo:

$$ y_e(t) = y(t) - y(t-1) $$

```python
import ipywidgets as widgets

def f(P=20,noise=0.5,trend=0.2):
  index_dates = pd.date_range(start='1/1/2000', end='1/1/2022', freq='M')
  values = np.sin( np.arange(0,len(index_dates),1) * np.pi/P )  +  np.cos( np.arange(0,len(index_dates),1) * np.pi/(2*P) ) + 10
  values = values + noise*np.random.sample(len(index_dates))
  values = values * trend*np.linspace(0,1,len(index_dates)) + 1

  values = values

  df = pd.DataFrame()
  df.index = index_dates 
  df['values'] = values

  df = df.diff().dropna() 
  plot_series(df['values'])  

  return df

w = widgets.interactive(f, P=(1, 100, 1), noise=(0, 1, 0.01), trend=(0,20,0.1))
display(w)

w.result
```

```python
from statsmodels.tsa.stattools import adfuller

df = w.result

X = df['values']
result = adfuller(X)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
	print('\t%s: %.3f' % (key, value))
```

Agora os dados se mostram, tanto graficamente como pelo teste ADF, estacionrios.
